```python
import random # This package provide a convenient way to generate the random number
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
seed = 42 # All the times get the same number
random.seed(seed)
```

```python
def random_Uniform(a,b,n):

  for i in range(n):
    random_numbers = [[random.random() * (b - a) + a for _ in range(n)]]
  return random_numbers
```

```python
a = 3
b = 5
n1 = 100
rand_unif_100 = random_Uniform(a,b,n1)
```

```python
n2 = 1000
rand_unif_1000 = random_Uniform(a,b,n2)
```

```python
plt.figure(figsize=(10, 5))  # Adjust figure size

plt.subplot(1, 2, 1)  # Subplot 1 (1 row, 2 columns, position 1)
plt.hist(rand_unif_100, bins=50, range=(a, b), color='blue', alpha=0.5, label='100', edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of 100 uniform random numbers')
plt.grid(True)

plt.subplot(1, 2, 2)  # Subplot 2 (1 row, 2 columns, position 2)
plt.hist(rand_unif_1000, bins=50, range=(a, b), color='orange', alpha=0.5, label='1000', edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of 1000 uniform random numbers')
plt.grid(True)


plt.tight_layout()
plt.show()
```

```python
def randomNormal(mean,std,n):
  mean = mean # Mean
  std_dev = std # Standard deviation

  normal_random_numbers = np.random.normal(mean, std_dev, n)
  return normal_random_numbers
```

```python
mean = 1
std = 2.5
n = 1000 # Size

# normal_100 = randomNormal(mean,std,n)
# n1 = 1000
normal_1000 = randomNormal(mean,std,n)
```

```python
sns.histplot(normal_1000, bins=20, kde=False, color='blue', alpha=0.5, label='100', edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of 1000 normal random numbers')
```
