```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def boltzmann_distribution(v, m, T):
    k = 1.38e-23  # Boltzmann constant in J/K
    return (4 * np.pi * v**2 / ((2 * np.pi * k * T) / m)**(3/2)) * np.exp(-m * v**2 / (2 * k * T))

# Parameters
temperature = 300  # Temperature in Kelvin
mass_of_molecule = 4.625e-26  # Mass of a nitrogen gas molecule in kg

# Speed range
speeds = np.linspace(0, 2000, 1000)  # Speeds from 0 to 2000 m/s

# Calculate Boltzmann distribution for each speed
probabilities = boltzmann_distribution(speeds, mass_of_molecule, temperature)

# Plotting
plt.plot(speeds, probabilities, label=f'Temperature = {temperature} K, Mass = {mass_of_molecule:.2e} kg')
plt.title('Boltzmann Distribution of Speeds')
plt.xlabel('Speed (m/s)')
plt.ylabel('Probability Density')
plt.legend()
plt.show()
```

```python
temp = [300,400,500,600,700]


for t in temp:
  f_values = boltzmann_distribution(speeds,mass_of_molecule,t)

  plt.plot(speeds, f_values, label='T = {} K'.format(t))

plt.xlabel('Speed (m/s)')
plt.ylabel('Probability Density')
plt.title('Boltzmann Distribution for Real-World Gas Molecules in Different Temperature')
plt.legend()
plt.grid(True)
plt.show()
```
