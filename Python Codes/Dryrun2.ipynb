```python
import numpy as np
import math
import random
import matplotlib.pyplot as plt

# Lists to store values from each iteration
current_r_values = []
ans_values = []
ite = []

i = 0

current_r = random.uniform(-1, 1)
# Iterate 100 times
for num in range(100):
    new_r = random.uniform(-1, 1)

    while current_r == new_r:
        #current_r = random.uniform(-1, 1)
        new_r = random.uniform(-1, 1)


    def energy_function(x):
        return x ** 2


    f_1 = energy_function(current_r)
    f_2 = energy_function(new_r)

    delta_e = f_2 - f_1

    if delta_e < 0:
        current_r = new_r
        #new_r = random.uniform(-1, 1)
    else:
        ans = math.exp(-1 * delta_e / 100)
        ans_r = random.uniform(0, 1)

        if ans > ans_r:
            current_r = new_r
        else:
            current_r = current_r
    i = i+1
        #print(current_r)
    # Store values for plotting
    current_r_values.append(current_r)
    ans_values.append(energy_function(current_r))
    ite.append(i)


    # Find the minimum value of current_r_values
min_current_r = min(current_r_values)

    # Find the minimum value of ans_values
#min_ans = min(ans_values)

print(f"Minimum value of current_r_values: {min_current_r}")
#print(f"Minimum value of ans_values: {min_ans}")

# Plot current_r vs ans
plt.figure(figsize=(8, 6))
# plt.scatter(ite,ans_values , label=f'Temperature = {100} K', color='blue', marker='o')
# plt.xlabel('iteration')
# plt.ylabel('ans')
# plt.title('Boltzmann Distribution Plot')
# plt.legend()
# plt.grid(True)
# plt.show()


plt.plot(ite,current_r_values, label=f'Temperature = {100} K', color='blue', marker='o')
plt.xlabel('iteration')
plt.ylabel('current')
plt.title('Boltzmann Distribution Plot')
plt.legend()
plt.grid(True)
plt.show()

plt.plot(ite,ans_values, label=f'Temperature = {100} K', color='blue', marker='o')
plt.xlabel('iteration')
plt.ylabel('Answervalue')
plt.title('Boltzmann Distribution Plot')
plt.legend()
plt.grid(True)
plt.show()
```
