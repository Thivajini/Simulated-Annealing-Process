```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
# Energy Function
def energy(param):
  return param**2
```

```python
def energydiff(state1,state2):
  return (energy(state2)-energy(state1))
```

```python
random.seed(123)
```

```python
def dryRun(n_iter):
    state_current = random.uniform(-1, 1)
    for i in range(n_iter):
        state_new = random.uniform(-1, 1)

        while state_new == state_current:
            state_new = random.uniform(-1, 1)

        delta_e = energydiff(state_current, state_new)  # Energy difference

        if delta_e < 0:
            state_current = state_new  # Set the current state as new state
        else:
            boltzman_factor = np.exp(-delta_e / 70)
            if boltzman_factor > random.uniform(0, 1):
                state_current = state_new
            else:
                state_current = state_current
    return state_current
```

```python
x = dryRun(100)
print("The current State: ",x)
print("Minimum Energy: ",energy(x))
```
